# Supplementary Material

```python
import pandas as pd
host_cities_data = pd.read_html("https://en.wikipedia.org/wiki/List_of_Olympic_Games_host_cities")
gdp_data = pd.read_csv("gdp.csv", skiprows=4)
population_data = pd.read_csv("population.csv", skiprows=4)
life_expetancy_data = pd.read_csv("life-expectancy.csv")
winter_data = pd.read_csv("winter.csv")
summer_data = pd.read_csv("summer.csv")
```

```python
dictionary_data = pd.read_csv("dictionary.csv")
dictionary_data = dictionary_data.drop(columns = ['Population', 'GDP per Capita'])
f = lambda x: x.replace('*', '')
dictionary_data['Country'] = dictionary_data['Country'].apply(f)
dictionary_data
```

```python
host_cities_data = host_cities_data[1].iloc[:, 2:4]
host_cities_data.drop(host_cities_data[host_cities_data.Year < 1960].index, inplace = True)
host_cities_data.drop(host_cities_data[host_cities_data.Year > 2020].index, inplace = True)
host_cities_data
```

```python
gdp_data = gdp_data.drop(columns = ["Country Code", "Indicator Name", "Indicator Code", "Unnamed: 65" ])
gdp_data
```

```python
gdp_data = gdp_data.T.fillna(gdp_data.mean(axis=1)).T
gdp_data
```

```python
gdp_data_pivot = pd.melt(gdp_data, id_vars = 'Country Name', var_name = 'Year', value_name = "GDP")
gdp_data_pivot.rename(columns = {'Country Name': 'Country'}, inplace = True)
gdp_data_pivot['Year'] = gdp_data_pivot['Year'].astype('int64')
gdp_data_pivot
```

```python
population_data = population_data.drop(columns = ["Country Code", "Indicator Name", "Indicator Code", "Unnamed: 65" ])
population_data
```

```python
population_data_pivot = pd.melt(population_data, id_vars = 'Country Name', var_name = 'Year', value_name = "Population")
population_data_pivot.rename(columns = {'Country Name': 'Country'}, inplace = True )
population_data_pivot['Year'] = population_data_pivot['Year'].astype('int64')
population_data_pivot
```

```python
life_expetancy_data = life_expetancy_data.drop(columns = 'Code')
life_expetancy_data.rename(columns = {'Entity': 'Country'}, inplace = True)
life_expetancy_data
```

```python
# count medals per country per year 
winter_data = winter_data.drop(columns = ['City', 'Sport', 'Discipline', 'Athlete', 'Gender', 'Event'])
winter_data
```

```python
winter = winter_data.copy()
winter['bronze']=(winter['Medal']=='Bronze').astype(int)
winter['silver']=(winter['Medal']=='Silver').astype(int)
winter['gold']=(winter['Medal']=='Gold').astype(int)
winter = winter.drop(columns = 'Medal')
```

```python
winter = winter.groupby(['Country', 'Year']).sum()
#winter.loc['USA']
#winter.loc['USA'].loc[1994]
winter.reset_index(inplace = True)
winter
```

```python
summer_data = summer_data.drop(columns = ['City', 'Sport', 'Discipline', 'Athlete', 'Gender', 'Event'])
```

```python
summer = summer_data.copy()
summer['bronze']=(summer['Medal']=='Bronze').astype(int)
summer['silver']=(summer['Medal']=='Silver').astype(int)
summer['gold']=(summer['Medal']=='Gold').astype(int)
summer = summer.drop(columns = 'Medal')
summer = summer.groupby(['Country', 'Year']).sum()
summer.reset_index(inplace = True)
summer
```

```python
medal_data = pd.concat([summer,winter])
medal_data = medal_data.merge(dictionary_data, left_on = 'Country', right_on = 'Code')
medal_data.drop(columns = ['Country_x', 'Code'], inplace = True)
medal_data = medal_data.rename(columns = {'Country_y': 'Country'})
medal_data = medal_data[['Country', 'Year', 'gold', 'silver', 'bronze']]
medal_data.drop(medal_data[medal_data.Year < 1960].index, inplace = True )
medal_data
```

## Host country vs medals

```python
host_cities_data.Country.unique()
```

```python
def is_host(country, year):
    for host in host_cities_data.values:
        if host[0] == country and host[1] == year: return True
    return False

def plot_host_medal_data(medal_data, season):
    # Replace country codes with country names
    medal_data = medal_data.merge(dictionary_data, left_on = 'Country', right_on = 'Code')
    medal_data.drop(columns = ['Country_x', 'Code'], inplace = True)
    medal_data = medal_data.rename(columns = {'Country_y': 'Country'})
    medal_data = medal_data[['Country', 'Year', 'gold', 'silver', 'bronze']]
    medal_data.drop(medal_data[medal_data.Year < 1960].index, inplace = True )
    if season == 'Summer':
        removed_summer_countries = ['Brazil','Norway','Russia']
        medal_data = medal_data[~medal_data['Country'].isin(removed_summer_countries)]
    else:
        removed_winter_countries = ['Australia','China','United Kingdom']
        medal_data = medal_data[~medal_data['Country'].isin(removed_winter_countries)]
    # Keep only medal data for countries that were host countries   
    host_medal_data = medal_data[medal_data.Country.isin(host_cities_data.Country.unique())]
    host_medal_data = host_medal_data.groupby(['Country','Year']).agg(['sum']).reset_index()
    host_medal_data.columns = ['Country','Year','gold','silver','bronze']
    # Create total medal count and total_host columns
    host_medal_data['total'] = host_medal_data['gold'] + host_medal_data['silver'] + host_medal_data['bronze']
    total_host = []
    for index, row in host_medal_data.iterrows():
        total_host.append(row['total'] if is_host(row['Country'], row['Year']) else None)
    host_medal_data['total_host'] = total_host
    # Plotting
    a = sns.FacetGrid(host_medal_data, col = 'Country',col_wrap=3, height = 10)
    a.map(sns.lineplot,'Year','total',alpha = 0.75, label = 'Total Medal Count', ci = 0)
    a.map(sns.scatterplot,'Year','total_host',alpha = 0.75, label='Medal Count as Host',color = 'red', s=100)
    for ax in a.axes.flat:
        ax.set_ylabel('Total Medal Count')
    a.add_legend()
    a.fig.subplots_adjust(top=.93)
    a.fig.suptitle('Total Medal Count for '+season+' Host Countries')
```

```python
plot_host_medal_data(summer, 'Summer')
plt.savefig('summer_host_medals.png')
```

```python
plot_host_medal_data(winter, 'Winter')
plt.savefig('winter_host_medals.png')
```
